```python
import pandas as pd
```

```python
#same dataset
rating = pd.read_csv("ratings.csv")
movie = pd.read_csv("movies_metadata.csv")
#our main movie dataset
main_movie = pd.read_csv("movies_metadata.csv")
```

```python
#dropping unnecessary columns, including only the title and id so it can be merged
rating.drop(columns=["timestamp"], inplace=True)
movie = movie[["original_title", "id"]]
```

```python
#dropping all NaN values
rating.dropna(inplace=True)
movie.dropna(inplace=True)
```

```python
rating["movieId"] = rating["movieId"].astype(str)
```

```python
#merging rating and movie datasets
combined_data = pd.merge(rating, movie, left_on="movieId", right_on="id", how="left")
```

```python
combined_data.dropna(inplace=True)
combined_data
```

```python
#dropping our redundant movieid/id column
combined_data.drop(columns=['id'], inplace=True)
```

```python
#renaming our 'original_title' column because it also exists in main_movie dataset
combined_data.rename(columns={'original_title': 'com_title'}, inplace=True)
combined_data
```

```python
#merge movies from our main movie dataset and combined dataset 
main_movie['title'] = main_movie['title'].str.strip().str.lower()
combined_data['com_title'] = combined_data['com_title'].str.strip().str.lower()

final_data = pd.merge(main_movie, combined_data, left_on='title', right_on='com_title', how='inner')
```

```python
#showing that merging indeed combined the same titles
final_data_new = final_data.drop_duplicates(subset='title')
final_data_new[['title', 'com_title']]
```

```python
#dropping our redundant com_title/title column
final_data.drop(columns=['com_title', 'original_title'], inplace=True)
final_data
```
